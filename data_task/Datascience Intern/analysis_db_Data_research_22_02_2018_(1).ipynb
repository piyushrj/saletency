We can clearly see that this is a database containing details of Indian companies.

### Grouping the companies by industry

```python
df.groupby(['Industry'])['Company Name'].count()
```

```python
df.groupby(['Industry'])['Company Name'].count().plot(kind='bar')
```

We can see that a large number of companies (37 to be precise) are **Information Technology (IT)** companies, which are much more than the next most popular **Engineering** companies.

### Grouping the companies by Type

```python
df.info()
```
